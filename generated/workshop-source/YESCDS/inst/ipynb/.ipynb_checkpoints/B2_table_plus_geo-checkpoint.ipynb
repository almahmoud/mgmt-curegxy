# The basic geographic data about US counties

The coordinates of central locations in US counties is provided in
`us_county_geo`.

```r
data(us_county_geo)
us_county_geo |> select(state, county, geometry)
```

Coordinates for metropolitan statistical areas are provided by
“statcrunch”:

```r
data(statcrunch_msa)
head(statcrunch_msa)
```

This geographic data has already been used to enhance the CDC incidence
data in `woncan`:

```r
data(woncan)
woncan |> group_by(MSA) |> 
    summarize(lat=head(lat)[1], lng=head(lng)[1]) |> head()
```
